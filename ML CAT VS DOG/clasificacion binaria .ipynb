```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from tensorflow.keras.metrics import Precision, Recall, AUC
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input
from PIL import Image, UnidentifiedImageError
import random
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
# Establecer la semilla para TensorFlow, NumPy y el generador de Python
seed = 42
tf.random.set_seed(seed)
np.random.seed(seed)
random.seed(seed)
```

```python
dataset_dir = "C:/Users/danny.valerio/Documents/Danny/Talleres/Autoeding/PetImages"  # Cambia esto a la ruta de tu carpeta "dataset"
```

```python

def validate_images(folder):
    valid_images = []
    for filename in os.listdir(folder):
        try:
            img = Image.open(os.path.join(folder, filename))
            img.verify()  # Verifica si el archivo es una imagen válida
            valid_images.append(filename)
        except (IOError, SyntaxError) as e:
            print(f"Archivo inválido o corrupto detectado: {filename} -> {e}")
    return valid_images

valid_files = validate_images(dataset_dir+'/dog')
print(f"Total de imágenes válidas: {len(valid_files)}")
```

```python
# Recorrer todas las imágenes en el directorio
for root, dirs, files in os.walk(dataset_dir):
    for file in files:
        
        if file.endswith('.tiff') or file.endswith('.tif'):  # Comprobar si es un archivo TIFF
            file_path = os.path.join(root, file)
            try:
                img = Image.open(file_path)  # Intentar abrir la imagen
                img.verify()  # Verifica si la imagen se puede abrir
                print(f"Archivo bueno encontrado : {file_path}")
            except (IOError, SyntaxError) as e:
                print(f"Archivo corrupto encontrado y eliminado: {file_path}")
                os.remove(file_path)  # Eliminar archivo corrupto
```

```python
# Crear el generador de datos con separación de entrenamiento y validación
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    horizontal_flip=True,  # Si usas transformaciones aleatorias
    
)
```

```python
# Cargar las imágenes de entrenamiento
train_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(150, 150),  # Ajusta el tamaño de las imágenes
    batch_size=50,           # Tamaño del batch
    class_mode='binary',     # Como es una clasificación binaria
    subset='training' ,       # Usa el conjunto de entrenamiento (80%)
    seed=seed  # Semilla para reproducibilidad
)
```

```python
# Cargar las imágenes de validación
validation_generator = datagen.flow_from_directory(
    dataset_dir,
    target_size=(150, 150),
    batch_size=50,
    class_mode='binary',
    subset='validation',      # Usa el conjunto de validación (20%)
    seed=seed  # Semilla para reproducibilidad
)
```

```python
# Crear el modelo de CNN
model = tf.keras.models.Sequential([
    Input(shape=(150, 150, 3)),  # Especificar el input con `Input` directamente
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.5, seed=seed),  # Establece la semilla aquí
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Clasificación binaria
])
```

```python
def f1_score(y_true, y_pred):
    # Convertir probabilidades a 0 o 1
    y_pred = K.round(y_pred)
    
    # Calcular precision
    precision = K.sum(K.round(K.clip(y_true * y_pred, 0, 1))) / (K.sum(K.round(K.clip(y_pred, 0, 1))) + K.epsilon())
    
    # Calcular recall
    recall = K.sum(K.round(K.clip(y_true * y_pred, 0, 1))) / (K.sum(K.round(K.clip(y_true, 0, 1))) + K.epsilon())
    
    # Calcular el F1
    f1 = 2 * (precision * recall) / (precision + recall + K.epsilon())
    return f1

def specificity(y_true, y_pred):
    # Convertir probabilidades a 0 o 1
    y_pred = K.round(y_pred)
    
    # Calcular verdaderos negativos
    true_negatives = K.sum(K.round(K.clip((1 - y_true) * (1 - y_pred), 0, 1)))
    
    # Calcular posibles negativos
    possible_negatives = K.sum(K.round(K.clip(1 - y_true, 0, 1)))
    
    # Calcular especificidad
    specificity = true_negatives / (possible_negatives + K.epsilon())
    return specificity
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy', Precision(), Recall(), f1_score, AUC(), specificity])

```

```python
print("Imágenes de entrenamiento:", train_generator.samples)
print("Imágenes de validación:", validation_generator.samples)
```

```python
print(f'Steps per epoch (train): {train_generator.samples // train_generator.batch_size}')
print(f'Validation steps: {validation_generator.samples // validation_generator.batch_size}')
```

```python
train_generator.reset()
validation_generator.reset()
history = model.fit(
    train_generator,
    steps_per_epoch= train_generator.samples // train_generator.batch_size,
    epochs=5,
    validation_data=validation_generator,
    validation_steps= validation_generator.samples // validation_generator.batch_size,
    #use_multiprocessing=True,  # Usa procesamiento en paralelo
    #workers=2  # Número de procesos paralelos
 )
```
