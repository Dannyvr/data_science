El presente proyecto se enfoca en aplicar técnicas de machine learning para realizar la segmentación de clientes y proporcionar una personalización efectiva a través de la plataforma del e-commerce. La segmentación de clientes es fundamental para comprender las necesidades, preferencias y comportamientos de los clientes, lo que permite adaptar las estrategias de marketing y mejorar la experiencia del cliente, de esta manera asegurarse de mejorar la retención de  los clientes. **Antecedentes**

La segmentación de clientes es un enfoque fundamental en el campo del marketing para comprender mejor las necesidades, preferencias y comportamientos de los clientes. Una segmentación efectiva permite a las empresas adaptar sus estrategias y ofrecer experiencias personalizadas que maximicen la satisfacción del cliente e impulsen el crecimiento del negocio. En este contexto, el algoritmo de clustering K-Means ha sido ampliamente utilizado y demostrado su eficacia en la segmentación de clientes.
Varios estudios previos han investigado el uso del algoritmo K-Means en la segmentación de clientes. Uno de esos estudios relevantes es el artículo titulado "Segmentación de clientes utilizando el algoritmo de clustering K-Means" (Vulchi, 2021). El objetivo principal de este artículo fue aplicar el algoritmo K-Means para agrupar a los clientes en segmentos con características similares, se destacan aspectos importantes como la inicialización de centroides y la asignación de puntos a los clusters correspondientes. 

“A Machine Learning Based Method for Customer Behavior Prediction”(Li et al. 2019) es otro artículo que menciona la utilidad de K-Means para la clasificación de clientes en clusters. Además menciona 2 métodos más que son bastantes utilizados a la hora de tratar de predecir el comportamiento de los clientes como lo son los “Decision Trees” y “Naive Bayes”. Y muestra los principales pasos a la hora de utilizarlos como lo son el preparar los datos, procesarlos, construir el modelo y aplicarlo.

Otra forma de segmentación de clientes para mejorar la experiencia y de esta manera evitar la pérdida de los mismos es el  Perceptrón multicapa.  Ismail et al. (2019) mencionan en su artículo “ Multi-Layer Perceptron Approach for Customer Churn Prediction” como se puede utilizar un perceptrón multicapa para predecir la pérdida de clientes y que este método de MLP da mejores resultado que los métodos tradicionales como lo es la regresión lineal.

**Metodo para graficar**

```python
# Evaluate model performance

def plot_confusion_matrix(y_test, predicted):    
    """
    Displays the confusion matrix and other metrics.
    
    Parameters:
      y_test : a list of test data. 
      predicted: a list of model results
    """
    classes = np.unique(y_test)
    
    ## Accuracy, Precision, Recall
    accuracy = metrics.accuracy_score(y_test, predicted)

    print("Accuracy:",  round(accuracy,2))
    print("Details:")
    print(metrics.classification_report(y_test, predicted))
    
    ## Plot confusion matrix
    cm = metrics.confusion_matrix(y_test, predicted)
    plt.figure(figsize=(10, 8))  # Ajustar el tamaño de la figura
    ax = plt.subplot()
    sns.heatmap(cm, annot=True, fmt='d', ax=ax, cmap=plt.cm.Blues, 
                cbar=False)
    ax.set(xlabel="Pred", ylabel="True", xticklabels=classes, 
           yticklabels=classes, title="Confusion matrix")
    plt.xticks(rotation=90)  # Girar las etiquetas del eje x
    plt.yticks(rotation=0)

    plt.show()
```

### Random Forest

```python
# get a list of models to evaluate
def get_models():
 models = dict()
 # define number of trees to consider
 n_trees = [10, 50, 100,200]
 for n in n_trees:
     models[str(n)] = RandomForestClassifier(n_estimators=n)
 return models
 
# evaluate a given model using cross-validation
def evaluate_model(model, X, y):
 # define the evaluation procedure
 cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
 # evaluate the model and collect the results
 scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
 return scores
```

```python
# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
    # evaluate the model
    scores = evaluate_model(model, X_train, y_train)
    # store the results
    results.append(scores)
    names.append(name)
    # summarize the performance along the way
    print("Número de árboles, precisión promedio y desviación promedio", '>%s %.3f (%.3f)' % (name, mean(scores), std(scores)))
```

**Funcion Relu**

```python
class MLPClassifier(nn.Module):
 
    def __init__(self, input_dim, hidden_dim1,hidden_dim2, output_dim):
       

        super(MLPClassifier, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
         
        self.fc2 = nn.Linear(hidden_dim, hidden_dim2)
        
        self.fc3 = nn.Linear(hidden_dim2, output_dim)
        
        self.dropout = nn.Dropout(0.1)


    def forward(self, x_in, apply_softmax=True):
        weight_dtype = self.fc1.weight.dtype

        # Convert x_in to the same dtype as self.fc1.weight
        x_in = x_in.to(weight_dtype)
        
        intermediate_vector = F.relu(self.fc1(x_in))
        intermediate_vector = self.dropout(intermediate_vector)
        
        intermediate_vector = F.relu(self.fc2(intermediate_vector))
        
        intermediate_vector = self.dropout(intermediate_vector)
        prediction_vector = self.fc3(intermediate_vector)
       
        if apply_softmax:
            prediction_vector = F.softmax(prediction_vector, dim=1)

        return prediction_vector
```

```python
# Model training

def model_training(epochs, classifier, loss_function, optimizer, X_train_tensor, y_train_tensor,
                   X_test_tensor, y_test_tensor):
    train_losses = []
    test_losses = []
    test_accuracies = []

    weight_dtype = classifier.fc1.weight.dtype

    for e in range(epochs):
        # Initialize the gradient
        optimizer.zero_grad()

        # Convert X_train_tensor to the same dtype as self.fc1.weight
        X_train_tensor = X_train_tensor.to(weight_dtype)

        # Model forward
        output = classifier.forward(X_train_tensor)

        # Compute the loss
        loss = loss_function(output, y_train_tensor.long())

        # Compute gradients
        loss.backward()

        # Update the train_losses vector
        train_loss = loss.item()
        train_losses.append(train_loss)

        optimizer.step()

        # Calculate model accuracy during training
        # Turn off gradients for test
        with torch.no_grad():
            # Model in eval mode
            classifier.eval()
            log_ps = classifier(X_test_tensor)
            test_loss = loss_function(log_ps, y_test_tensor)
            test_losses.append(test_loss)

            # Accuracy
            # topk: Returns the k largest elements of the given
            #     input tensor along a given dimension.
            top_p, top_class = log_ps.topk(1, dim=1)
            equals = top_class == y_test_tensor.view(*top_class.shape)
            test_accuracy = torch.mean(equals.float())
            test_accuracies.append(test_accuracy)

        classifier.train()
        if (e+1) % 100 == 0:
            print(f"Epoch: {e+1}/{epochs}.. ",
                  f"Test Accuracy: {test_accuracy:.3f}")

    return classifier, train_losses, test_losses, test_accuracies

        
# Model definition

epochs = 6000
```

```python
classifier, train_losses, test_losses, test_accuracies =  model_training(epochs, classifier, 
                                                                         loss_function, 
                                                                         optimizer, 
                                                                         X_train_tensor, 
                                                                         y_train_tensor, 
                                                                         X_test_tensor, y_test_tensor)
```

```python
## Loss curve

def loss_curve (train_losses,test_losses, test_accuracies, y_label,start):
    """ 
    Displays the loss curve.
    Parameters: 
       train_losses: list of loss values per epoch using training data.
       test_losses: list of loss values per epoch using test data.
       test_accuracies: : a list of accuracie by epoch using test data. 
       y_label
    """ 
    
    train_losses = train_losses[start:]
    test_losses = test_losses[start:]
    test_accuracies = test_accuracies[start:]
    
    
    plt.figure(figsize=(12, 5))
    ax = plt.subplot(121)
    plt.xlabel('epochs')
    plt.ylabel(y_label)
    plt.plot(train_losses, label='Training loss')
    plt.plot(test_losses, label='Test loss')
    plt.legend(frameon=False);
    plt.subplot(122)
    plt.xlabel('epochs')
    plt.ylabel('test accuracy')
    plt.plot(test_accuracies)

    
    
loss_curve (train_losses,test_losses, test_accuracies, 'negative log likelihood loss',0)
```
