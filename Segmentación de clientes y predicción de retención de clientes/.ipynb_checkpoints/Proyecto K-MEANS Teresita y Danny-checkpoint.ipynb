# IMPORTAR LAS LIBRERÍAS Y LOS DATASETS

```python
##K-Means es el algoritmo de agrupación en clústeres más conocido e implica la asignación de muestras a los clústeres en un 
##esfuerzo por minimizar la varianza dentro de cada clúster. Para la asignación de utilizan medidas de distancia
##y en cada iteración se asigna cada muestra al centroide más cercano.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import category_encoders as ce
```

# Carga de datos

```python
df = pd.read_csv('myData.csv', sep=',',encoding='utf-8')
df.shape
```

```python
columnas =['Country', 'NEW','Ages', 'Time', 'Total__service1', 'Total__service2',
       'Total_items_service1','Total_Items_service2', 'Amount_service1', 'Amount_service2', 'Taxes__service1','Taxes_service2', 
       'Cancellations_service1', 'Cancellations_service2',        
       'Type_items_1', 'Tyep_items_2', 'CHURN']
df=df[columnas]
```

```python
df.head(10)
```

# Preprocesamiento

```python
df.describe()
```

# Limpiar las edades

```python
df[df['Ages']<0].shape
```

```python
df.loc[df['Ages'] < 0, 'Ages'] = 0
```

## En el caso de los tiempos y Montos eliminamos las filas

```python
# Seleccionar las filas que cumplen la condición
condicion = (df['Time'] < 0) | (df['Amount_service2'] < 0)
filas_seleccionadas = df[condicion]
filas_seleccionadas.shape
```

```python
# Eliminar las filas seleccionadas
df = df.drop(filas_seleccionadas.index)

# Imprimir el DataFrame resultante
df.shape
```

```python
#Valdiar datos faltantes
sns.heatmap(df.isnull(), yticklabels=False, cbar = False, cmap = "Blues")
```

# codificación de datos categóricos

```python
print('Country:',df['Country'].unique())
```

```python
counts = df['Country'].value_counts()
print(counts)
```

```python
#Aplicaremos un metodo binning (principio de Pareto o regla del 80-20) para reducir los paises ya que muchos estan en pocas categorias, pasando asi a solo 8
```

```python
other_vals = counts[counts < 8010].index.tolist()
# Reemplazar los valores en la columna categórica con "otros"
df.loc[df['Country'].isin(other_vals), 'Country'] = 'other'

counts = df['Country'].value_counts()
print('---')
print('despues')
print(counts)
```

```python
# An instance of the encoder is created 
encoder=ce.sum_coding.SumEncoder(cols=['Country'],verbose=False)
df=encoder.fit_transform(df)

df=df.drop(['intercept'], axis=1)
df
```

```python
# Empecemos por escalar primero el dataset
scaler = StandardScaler()
df_scaled  = scaler.fit_transform(df)
```

```python
df_scaled .shape
```

```python
df_scaled 
```

# EL NÚMERO ÓPTIMO DE CLISTERS UTILIZANDO EL MÉTODO DEL CODO

```python
scores_1 = []

range_values = range(1, 20)

for i in range_values:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_scaled )
    scores_1.append(kmeans.inertia_) #WCSS


plt.plot(range_values, scores_1, 'bx-')
plt.title("Encontrar el número óptimo de Clusters")
plt.xlabel("Clusters")
plt.ylabel("WCSS(k)")
plt.show()




# Sin embargo, los valores se reducen a una forma lineal hasta el 5º cluster. 
# Se elije el número de clusters igual a 5.
```

# APLICAR EL MÉTODO DE K-MEANS

```python
kmeans = KMeans(5)
kmeans.fit(df_scaled )
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns=[df.columns]) 
cluster_centers
```

```python
# Para entender mejor estos valores, vamos a aplicar la transformación inversa.
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns=[df.columns])
cluster_centers

#Cluster 0  Desertores servicio 2  Altas compras en el servicios 2 pero con altas cancelaciones, tiempo medio de ser registrados como clientes.
#Cluester 1 Nuevo en servicio 1 con alto  tiempo de espera: Clientes recien registrados que tiene una alta duracion en recibir el pedido.
#Cluester 2 Clientes unicos para el servicio 1 con alto nivel de adquisicion
#Cluster 3 Clientes constantes en el servicio 1 bajo nivel de cancelacion
#Cluster 4: Clientes con alta antiguedad de registro y bajo nivel de gatos servicio 1
```

```python
cluster_centers[['Ages','Time','Total__service1', 'Total__service2','Total_items_service1','Total_items_service1','Cancellations_service1','Cancellations_service2','Amount_service1','Amount_service2']]
```

```python
labels.shape
```

```python
y_kmeans = kmeans.fit_predict(df_scaled )
y_kmeans
```

```python
# Concatenamos las etiquetas de los clusters con el dataset riginal 
df_cluster = pd.concat([df, pd.DataFrame({'cluster': labels})], axis = 1)
df_cluster.head()
```

```python
# Visualizamos histogramas para cada cluster 
for i in df.columns:
    plt.figure(figsize=(35, 5))
    for j in range(8):
        plt.subplot(1, 8, j+1)
        cluster = df_cluster[df_cluster['cluster'] == j]
        cluster[i].hist(bins = 20)
        plt.title('{}    \nCluster {}'.format(i, j))
    plt.show()
```

# ANÁLISIS DE LAS COMPONENTES PRINCIPALES

```python
# Obtener las componentes principales
pca = PCA(n_components = 2)
principal_comp = pca.fit_transform(df_scaled)
principal_comp
```

```python
# Crear un dataframe con las dos componentes 
pca_df = pd.DataFrame(data = principal_comp, columns=["pca1", "pca2"])
pca_df.head()
```

```python
# Concater las etiquetas de los clusters con el dataframe de las componentes principales

pca_df = pd.concat([pca_df, pd.DataFrame({'cluster':labels})], axis = 1)

pca_df.head()
```

```python
plt.figure(figsize=(8,8))
ax = sns.scatterplot(x = "pca1", y = "pca2", hue = "cluster", data = pca_df,
                     palette = ["red", "green", "blue", "pink", "yellow", "gray", "purple", "black"])
plt.show()
```
