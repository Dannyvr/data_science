```python
def binarize_image(image_tensor):
    image_tensor[image_tensor > 0.5] = 1
    image_tensor[image_tensor <= 0.5] = 0
    return image_tensor




def load_dataset(path = "mnist_dataset/train"):
    #Open up the dataset
    dataset =  torchvision.datasets.ImageFolder(path)   
    #print(dataset)
    list_images = dataset.imgs
    #print(list_images)
    train_data_tensor  = None    
    labels_training = []
    first_tensor = True
    #list_images_training =  set(data_labeled.train_ds.x.items)
    #print(list_images)
    for i in range(0, len(list_images)):
        pair_path_label = list_images[i]        
        image = Image.open(pair_path_label[0])        
        #print("path ", pair_path_label[0])
        x_tensor = TF.to_tensor(image).squeeze()
        x_tensor_bin = binarize_image(x_tensor)
        #print("x_tensor_bin")
        #print(x_tensor_bin.shape)
        #plt.figure()
        #plt.imshow(x_tensor_bin)
        x_tensor_bin_plain = x_tensor_bin.view(x_tensor_bin.shape[0] * x_tensor_bin.shape[1], -1)
        #print("x_tensor_bin_plain")
        #print(x_tensor_bin_plain.shape)
        #print("tensor ", x_tensor.shape)
        #test dataset case        
        #if("train" in pair_path_label[0]):
        labels_training += [pair_path_label[1]]
        #print(pair_path_label)
        if(first_tensor):
            first_tensor = False
            train_data_tensor = x_tensor_bin_plain
        else:
            train_data_tensor = torch.cat((train_data_tensor, x_tensor_bin_plain), 1)    
    return (train_data_tensor, torch.tensor(labels_training))       

```

## Función train_model

```python
def train_model(train_data_tensor, labels_training, num_classes = 10):
    # calcularmos el priori de  p(t) para clases K
    _, conteo = torch.unique(labels_training, return_counts=True)
    p_k=conteo/conteo.sum()
    #print(p_k)

    #caluclamos la probabilidade por pixel
    total = train_data_tensor.shape[1]
    p_m_1_given_k = torch.tensor([])

    for k in conteo:
        # escoge el numero de imagenes de la clase k
        
        k_m = train_data_tensor[:, :k]
        # suma los pixeles 1 para saber cuantos hay
        k_m = torch.sum(k_m, dim=1)
        #print(k_m)
        
         # calcula la probavilidad de que los  pixeles sean 1
        k_m = k_m / k       
        k_m=k_m.unsqueeze(1)
        #print(k_m)
        train_data_tensor = train_data_tensor[:, k:]
        p_m_1_given_k = torch.cat((p_m_1_given_k, k_m), dim=1)   #concatenar los tensores de probabilidades por imagen

        #print(p_m_1_k.shape)

    return (p_m_1_given_k ,p_k)
```

```python
def test_model(input_torch, p_m_pix_val_given_k, p_t_tensor, num_classes = 10):
    
    
    vector =input_torch
    vector=vector.unsqueeze(1)

    #print('vector',vector)
    matriz = p_m_pix_val_given_k
    #print('matriz',matriz)
    
    #calcula el vector ocupues para la probabilidad de que sea 0
    nuevo_vector = 1-vector
    
    #sacamos la priubabilidad de cada pixel
    resultado1 = vector*matriz + nuevo_vector*(1-matriz)
    #print('resultado',resultado1)
    
    # creamos un valor minimo para evitar infinitos, indefinidos o nulos
    e=0.00000000001
    #multiplicamos por la probabilidad priore de K
    #print(resultado1)
    #print(p_t_tensor)
    resultado2=torch.log(resultado1+e)
    #print(resultado2)
    #scores_classes = torch.prod(resultado2, dim=0)
    # sacamos log y sumanos para evitar 0 con la multiplicaciones
    resultado3=torch.sum(resultado2, dim=0)   
    scores_classes = resultado3*p_t_tensor 
    
    #print(scores_classes)
    predicted_label = torch.argmax(scores_classes)
    #print(predicted_label)
    
    
    return (predicted_label, scores_classes)
```

```python
def test_model_batch(test_set, labels, p_m_pix_val_given_k, p_t_tensor):
    num_columnas = test_set.shape[1]
 
    # Recorrer por columnas
    correctas = 0
    for columna in range(num_columnas):
        # Obtener la columna actual
        imagen = test_set[:, columna]
        #print(imagen)
        (predicted_label, scores_classes) = test_model(imagen, p_m_pix_val_given_k, p_t_tensor)
        #print("predicted_label ", predicted_label)
        #print("real label ", labels[columna])
        #print("scores_classes ", scores_classes)
        if labels[columna] == predicted_label:
            correctas += 1

    accuracy = correctas / num_columnas
    return accuracy
```

```python
def calculate_statistics_per_pixel(train_data_tensor, labels_training, num_classes=10):
    num_pixels = train_data_tensor.shape[0]
    num_samples = train_data_tensor.shape[1]

    # Inicializar listas para almacenar los datos
    means_per_pixel = torch.zeros((num_pixels, num_classes))
    medians_per_pixel = torch.zeros((num_pixels, num_classes))
    variances_per_pixel = torch.zeros((num_pixels, num_classes))

    # Calcular las estadísticas para cada píxel
    for k in range(num_classes):
        class_data = train_data_tensor[:, labels_training == k]

        for pixel in range(num_pixels):
            
            pixel_values = class_data[pixel]

            # Calcular la media, mediana y varianza para el píxel
            mean = torch.mean(pixel_values)
            median = torch.median(pixel_values)
            variance = torch.var(pixel_values)

            means_per_pixel[pixel, k] = mean
            medians_per_pixel[pixel, k] = median
            variances_per_pixel[pixel, k] = variance

    return means_per_pixel, medians_per_pixel, variances_per_pixel
```

## Modelos KernelDensity

```python
def train_model(train_data_tensor, labels_training,modelo, num_classes = 10):
    
    _, conteo = torch.unique(labels_training, return_counts=True)
    num_pixels = train_data_tensor.shape[0]
    p_k=conteo/conteo.sum()


    # la lista para almacenar las funciones de densidad de los píxeles para cada clase
    density_models = [[] for _ in range(num_classes)]

    # Calcular la función de densidad para cada pixel y cada clase
    #k=0
    for k in range(num_classes):
        class_data = train_data_tensor[:,labels_training == k]
        #print(k)
        #pixel=0
        for pixel in range(num_pixels):
            # Obtener los valores del píxel correspondiente a todas las imágenes de la clase k
            pixel_values = class_data[pixel]
            #print(pixel_values.unsqueeze(1))
            #print(pixel_values.shape)
            # Estimar la función de densidad para el píxel actual utilizando KDE
           
            #gaussian
            #exponential
            #epanechnikov
            #tophat
            kde = KernelDensity(kernel=modelo, bandwidth=1.0)
            
            kde.fit(pixel_values.unsqueeze(1))
            density_models[k].append(kde)
            #print(density_models)
            
    return density_models,p_k
```

```python
def test_model(input_data, density_models,p_t_tensor):
    num_classes = len(density_models)
    num_pixels = input_data.shape[0]

  
    # Calcular la probabilidad posterior para cada clase
    log_likelihoods = torch.zeros((input_data.shape[0], num_classes))
    
    for k in range(num_classes):
        
        for pixel in range(num_pixels):
            # Obtener los valores del píxel actual para todas las imágenes
            
            pixel_values = input_data[pixel]
            #print('valor: ',pixel_values)
            # Calcular la log-probabilidad utilizando la función de densidad KDE del pixel actual
            p=torch.tensor(density_models[k][pixel].score_samples(pixel_values.reshape(-1, 1)))
            
            log_likelihoods[:, k] += p
               
    resultado=log_likelihoods*p_t_tensor
    resultado2 = torch.sum(resultado, dim=0)

    # La clase predicha será la clase con la mayor probabilidad posterior
    predicted_labels = torch.argmax(resultado2)

    return predicted_labels,scores_classes
```

```python
def test_kernel_batch(test_set, labels, density_models,p_t_tensor):
    num_columnas = test_set.shape[1]
 
    # Recorrer por columnas
    correctas = 0
    for columna in range(num_columnas):
       
        imagen = test_set[:, columna]
        print(columna)
        (predicted_label, scores_classes) = test_model(imagen, density_models,p_t_tensor)
        #print("predicted_label ", predicted_label)
        #print("real label ", labels[columna])
        #print("scores_classes ", scores_classes)
        if labels[columna] == predicted_label:
            correctas += 1

    accuracy = correctas / num_columnas
    return accuracy
```

## Modelo Gausiano

```python
def train_model_gaussiano(train_data_tensor, labels_training, num_classes=10):
    _, conteo = torch.unique(labels_training, return_counts=True)
    num_pixels = train_data_tensor.shape[0]
    p_k = conteo / conteo.sum()

    # Convertir los datos y etiquetas a arreglos de numpy
    train_data_np = train_data_tensor.numpy()
    labels_training_np = labels_training.numpy()

    # La lista para almacenar las funciones de densidad gaussiana de los píxeles para cada clase
    density_models = [[] for _ in range(num_classes)]

    # Calcular la función de densidad gaussiana para cada pixel y cada clase
    for k in range(num_classes):
        class_data = train_data_np[:, labels_training_np == k]

        for pixel in range(num_pixels):
            # Obtener los valores del píxel correspondiente a todas las imágenes de la clase k
            pixel_values = class_data[pixel]

            # Ajustar la función de densidad gaussiana a los datos del píxel actual utilizando Máxima Verosimilitud
            mean, std = norm.fit(pixel_values)

            # Almacenar los parámetros de la función de densidad gaussiana en la lista
            density_models[k].append((mean, std))

    return density_models, p_k


def test_model_gaussiano(input_data, density_models, p_t_tensor):
    num_classes = len(density_models)
    num_pixels = input_data.shape[0]

    # Convertir los datos de prueba a un arreglo de numpy
    input_data_np = input_data.numpy()

    # Calcular la probabilidad posterior para cada clase
    log_likelihoods = torch.zeros((input_data.shape[0], num_classes))

    for k in range(num_classes):
        for pixel in range(num_pixels):
            # Obtener los valores del píxel actual para todas las imágenes de la clase k
            pixel_values = input_data_np[pixel]

            # Obtener los parámetros de la función de densidad gaussiana para el píxel actual y la clase k
            mean, std = density_models[k][pixel]

            epsilon = 1e-10
            #pixel_values = np.maximum(pixel_values, epsilon)

            # 
            p = np.log(norm.pdf(pixel_values, mean, std)+epsilon)
            log_likelihoods[:, k] += torch.tensor(p)

    result = log_likelihoods * p_t_tensor
    result_sum = torch.sum(result, dim=0)

    # La clase predicha será la clase con la mayor probabilidad posterior
    predicted_labels = torch.argmax(result_sum)

    return predicted_labels, scores_classes
```

```python
def test_gaussiano_batch(test_set, labels, density_models,p_t_tensor):
    num_columnas = test_set.shape[1]
 
    # Recorrer por columnas
    correctas = 0
    for columna in range(num_columnas):
        # Obtener la columna actual
        imagen = test_set[:, columna]
        print(columna)
        (predicted_label, scores_classes) = test_model_gaussiano(imagen, density_models,p_t_tensor)
        #print("predicted_label ", predicted_label)
        #print("real label ", labels[columna])
        #print("scores_classes ", scores_classes)
        if labels[columna] == predicted_label:
            correctas += 1

    accuracy = correctas / num_columnas
    return accuracy
```

## Modelo Weibull 

```python
def train_model_Weibull (train_data_tensor, labels_training, num_classes=10):
    _, conteo = torch.unique(labels_training, return_counts=True)
    num_pixels = train_data_tensor.shape[0]
    p_k = conteo / conteo.sum()

    # Convertir los datos y etiquetas a arreglos de numpy
    train_data_np = train_data_tensor.numpy()
    labels_training_np = labels_training.numpy()

    # Inicializar la lista para almacenar las funciones de densidad de Weibull de los píxeles para cada clase
    density_models = [[] for _ in range(num_classes)]

    # Calcular la función de densidad de Weibull para cada pixel y cada clase
    for k in range(num_classes):
        class_data = train_data_np[:, labels_training_np == k]

        for pixel in range(num_pixels):
            # Obtener los valores del píxel correspondiente a todas las imágenes de la clase k
            pixel_values = class_data[pixel]

            # Ajustar la función de densidad de Weibull a los datos del píxel actual utilizando Máxima Verosimilitud
            shape, loc, scale = weibull_min.fit(pixel_values, floc=0)

            # Almacenar los parámetros de la función de densidad de Weibull en la lista
            density_models[k].append((shape, loc, scale))

    return density_models, p_k


def test_model_Weibull (input_data, density_models, p_t_tensor):
    num_classes = len(density_models)
    num_pixels = input_data.shape[0]

    # Convertir los datos de prueba a un arreglo de numpy
    input_data_np = input_data.numpy()

    # Calcular la probabilidad posterior para cada clase
    log_likelihoods = torch.zeros((input_data.shape[0], num_classes))

    for k in range(num_classes):
        for pixel in range(num_pixels):
            # Obtener los valores del píxel actual para todas las imágenes de la clase k
            pixel_values = input_data_np[pixel]

            # Obtener los parámetros de la función de densidad de Weibull para el píxel actual y la clase k
            shape, loc, scale = density_models[k][pixel]
            epsilon = 1e-10
            # Calcular la log-probabilidad utilizando la función de densidad de Weibull
            p = np.log(weibull_min.pdf(pixel_values, shape, loc, scale)+epsilon)
            log_likelihoods[:, k] += torch.tensor(p)

    result = log_likelihoods * p_t_tensor
    result_sum = torch.sum(result, dim=0)

    # La clase predicha será la clase con la mayor probabilidad posterior
    predicted_labels = torch.argmax(result_sum)

    return predicted_labels, scores_classes
```

```python
def test_Weibull_batch(test_set, labels, density_models,p_t_tensor):
    num_columnas = test_set.shape[1]
 
    # Recorrer por columnas
    correctas = 0
    for columna in range(num_columnas):
        # Obtener la columna actual
        imagen = test_set[:, columna]
        print(columna)
        (predicted_label, scores_classes) = test_model_Weibull(imagen, density_models,p_t_tensor)
        #print("predicted_label ", predicted_label)
        #print("real label ", labels[columna])
        #print("scores_classes ", scores_classes)
        if labels[columna] == predicted_label:
            correctas += 1

    accuracy = correctas / num_columnas
    return accuracy
```
