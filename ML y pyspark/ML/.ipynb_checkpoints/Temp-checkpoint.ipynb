```python
conteo_por_riesgo = df['Heart Attack Risk'].value_counts()

# Crea un gráfico de barras
plt.figure(figsize=(8, 6))
ax = conteo_por_riesgo.plot(kind='bar')
plt.title('Número de Pacientes por Riesgo de Ataque Cardíaco')
plt.xlabel('Riesgo de Ataque Cardíaco')
plt.ylabel('Número de Pacientes')
plt.xticks(rotation=45)

# Agrega etiquetas en las barras
for i, v in enumerate(conteo_por_riesgo):
    ax.text(i, v + 10, str(v), ha='center', va='bottom', fontsize=10)

# Muestra el gráfico
plt.tight_layout()
plt.show()
```
